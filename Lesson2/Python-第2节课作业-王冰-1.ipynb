````python
y = input('input year: ')
y = int(y)
if ((0 == y % 4) and not (0 == y % 100)) or (0 == y % 400):
    print('the year {} is leap year'.format(y))
else:
    print('the year {} is not leap year'.format(y))
````

## 1. 实现一周的记账功能
首先记录输入的7天的收入，然后记录输入的7天的支出。  
然后打印出以下内容：
- 7天的收入(list)
- 7天的支出(list)
- 每天的结余(dict)
- 最终的结余

````python
in0   = []
out0  = []
save0 = dict()

for d in range(7):
    i = round(float(input('please input income for day{}'.format(d))),2)
    in0.append(i)
    
for d in range(7):
    i = round(float(input('please input outcome for day{}'.format(d))),2)
    out0.append(i)

for d in range(7):
    s = round((in0[d] - out0[d]),2)
    save0[d] = s
    
total = sum(in0) - sum(out0)

print('7天的收入(list) ===================================')
for i,v in enumerate(in0):
    print("day{}'s income: {}".format(i, v))

print('7天的支出(list) ===================================')
for i,v in enumerate(out0):
    print("day{}'s outcome: {}".format(i, v))

print('每天的结余(dict) ===================================')
for i,v in save0.items():
    print("day{}'s balance: {}".format(i, v))

print('最终的结余 =========================================')
print("final balance: {}".format(total))
````

## 点评

很棒，正确的实现了一周记账功能！使用了格式化输出！

在获取每日收入和支出时不够美观，中间加个冒号就好了。

## 2. 实现ATM机器
- 查询余额
- 存款
- 取款
- 退出

````python
display = """
===========
Please select
1. check account balance
2. account deposit
3. account withdraw
4. exit
===========

"""
bal = 0
cmd = 0

while 4 != cmd :
    cmd = int(input(display))
    
    if   1 == cmd :
        print('your balance is: {}'.format(bal))
    elif 2 == cmd :
        dep = input('how much do you want to deposit?')
        bal += round(float(dep),2)
        print('done. you deposit {} to your accont and current balance is: {}'.format(dep, bal))
    elif 3 == cmd :
        wrd = input('how much do you want to withdraw?')
        bal -= round(float(wrd),2)
        print('done. you withdraw {} from your accont and current balance is: {}'.format(wrd, bal))
    else:
        break
````
